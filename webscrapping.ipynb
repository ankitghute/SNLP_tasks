## Parsing the HTML content

```python
import requests 
from bs4 import BeautifulSoup 
  
URL = "https://towardsdatascience.com/named-entity-recognition-applications-and-use-cases-acdbf57d595e"
r = requests.get(URL) 
  
soup = BeautifulSoup(r.content, 'html.parser') 
print(soup.prettify()) 
```

```python
print(soup.title)
print(soup.p)
print(soup.find_all('a'))
```

```python
p_result = soup.find(id='714b')
print(p_result)
```

```python
print(p_result.prettify())
```

```python
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text)
```

```python
processed_text = remove_html_tags(str(p_result))
```

```python
processed_text
```

```python
fp1 = nlp(processed_text)
fp1
```

```python
# Tokenization 
for token in fp1:
    print (token, token.idx, end=" | ")
```

```python
# Removing stop words
for token in fp1:
    if not token.is_stop:
        print (token, end=" | ")
```

```python
# Lemmatization
for token in fp1:
    print (token, token.lemma_, end=" | ")
```
